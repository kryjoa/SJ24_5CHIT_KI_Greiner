## 2.1

```python
with open('Cyber-Sicherheit.pdf', 'rb') as pdf:
    reader = pypdf.PdfReader(pdf)

    metadata = reader.metadata

    author = metadata.get('/Author', 'Unbekannt')
    producer = metadata.get('/Producer', 'Unbekannt')
    title = metadata.get('/Title', 'Unbekannt')

    output = f"Author: {author}\nProducer: {producer}\nTitle: {title}"
    print(output)
```

```python
with open('Cyber-Sicherheit.pdf', 'rb') as pdf:
    reader = pypdf.PdfReader(pdf)

    first_page = reader.pages[0]
    first_page_text = first_page.extract_text()

    print(first_page_text)
```

```python
with open('Cyber-Sicherheit.pdf', 'rb') as pdf:
    reader = pypdf.PdfReader(pdf)

    all_pages = reader.pages
    
    for pages in all_pages:
        print(pages.extract_text())
```

```python
with open('Cyber-Sicherheit.pdf', 'rb') as pdf:
    reader = pypdf.PdfReader(pdf)

    all_pages = reader.pages 
    all_pages_list = []
    
    for pages in all_pages:
        all_pages_list.append(pages.extract_text())

    print(all_pages_list)
```

# Writing docx Word Files

```python
from docx import Document
from docx.shared import Inches 

document = Document()
document.add_heading('Document Title', 0)
p = document.add_paragraph('A plain paragraph having some ')
p.add_run('bold').bold = True
p.add_run(' and some ')
p.add_run('italic.').italic = True
document.add_heading('Heading, level 1', level=1)
document.add_paragraph('Intense quote', style='Intense Quote')
document.add_paragraph(
 'first item in unordered list', style='List Bullet'
)
document.add_paragraph(
 'first item in ordered list', style='List Number'
)
records = (
 (3, '101', 'Spam'),
 (7, '422', 'Eggs'),
 (4, '631', 'Spam, spam, eggs, and spam')
)
table = document.add_table(rows=1, cols=3)
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Qty'
hdr_cells[1].text = 'Id'
hdr_cells[2].text = 'Desc'
for qty, id, desc in records:
 row_cells = table.add_row().cells
 row_cells[0].text = str(qty)
 row_cells[1].text = id
 row_cells[2].text = desc
document.add_page_break()
document.save('demo.docx')
```

# Collecting Data from HTML


```python
# Bibliotheken laden
import urllib.request as urllib2
from bs4 import BeautifulSoup

# Website laden
response = urllib2.urlopen('https://www.htlkrems.ac.at')
html_doc = response.read()

# Das BeautifulSoup Object soup repräsentiert das „geparste“ HTML-Dokument
soup = BeautifulSoup(html_doc, 'html.parser')

# Das „geparste“ HTML-Dokument formatieren, sodass jeder Tag bzw. Textblock
# in einer separaten Zeile ausgegeben wird
strhtm = soup.prettify()

# Ein paar Zeilen ausgeben
print (strhtm[:1000]) 
```
