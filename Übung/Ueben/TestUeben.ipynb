```python
import spacy
from collections import Counter

nlp = spacy.load('en_core_web_sm')

with open('Text.txt', 'r', encoding='utf-8') as file:
    text = file.read()

doc = nlp(text)

words = []

for word in doc:
    if not word.is_stop and not word.is_punct:
        words.append(word.pos_)

counted = Counter(words)

print(counted)
```

schwieriges Beispiel: 
lese ein PDF ein, ersetze alle Personen mit Redacted und alle Organisationen mit beidl. 
dabei sollen alle punkte und stopwörter gelöscht werden und Verben sollen alle durch "JOH_DER_HUND" ersetzt werden!

```python
import spacy
import PyPDF2 as pypdf

nlp = spacy.load("de_core_news_sm")

with open("Bericht.pdf", "rb") as file:
    reader = pypdf.PdfReader(file)

    text = ""

    for page in reader.pages:
        text += page.extract_text()

doc = nlp(text)

redactedText = text

for ent in doc.ents:
    if ent.label_ == "PER":
        redactedText = redactedText.replace(ent.text, "[REDACTED]")
    if ent.label_ == "ORG":
        redactedText = redactedText.replace(ent.text, "beidl")

doc2 = nlp(redactedText)

words = []

for word in doc2:
    if not word.is_stop and not word.is_punct:
        if word.pos_ == "VERB":
            words.append("[JOH_DER_HUND]")
        else:
            words.append(word.text)

print(" ".join(words))
```
