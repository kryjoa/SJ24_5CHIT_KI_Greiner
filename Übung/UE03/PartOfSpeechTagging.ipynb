```python
# POS-Tagging mit NLTK
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

# NLTK-Ressourcen 
nltk.download('punkt')
nltk.download('tagsets')

text = "I love NLP and I will learn NLP in 2 month"
stop_words = set(stopwords.words('english')) 
words = [token for token in word_tokenize(text) if token not in stop_words] 

print(nltk.pos_tag(words))

print("------------")

nltk.help.upenn_tagset() 
```

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "I love NLP and I will learn NLP in 2 month"

doc = nlp(text)

words = [token for token in doc if token.is_stop == False]
for token in words:
    print(token.text, token.pos_)
```

```python
#3.2.1
import spacy

nlp = spacy.load("en_core_web_sm")

with open('dataScientistJobDesc.txt', 'r', encoding='utf-8') as file:
    text = file.read()

doc = nlp(text)

nouns_and_propn = [token.text for token in doc if token.pos_ in ['NOUN', 'PROPN']]

num_elements = len(nouns_and_propn)

print("Anzahl spacy:", num_elements)

```

```python
#3.2.2
import spacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")

with open('dataScientistJobDesc.txt', 'r', encoding='utf-8') as file:
    text = file.read()

doc = nlp(text)

pos_counts = Counter([token.pos_ for token in doc])

print(pos_counts)


#--------------------------------------------------
import matplotlib.pyplot as plt

pos_tags = list(pos_counts.keys())
frequencies = list(pos_counts.values())

plt.bar(pos_tags, frequencies, color='blue')

plt.xlabel('POS-Tags')
plt.ylabel('Häufigkeit')
plt.title('Häufigkeitsverteilung der Wortklassen')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#3.2.3
import spacy
from collections import Counter

nlp = spacy.load("en_core_web_sm")

with open('dataScientistJobDesc.txt', 'r', encoding='utf-8') as file:
    text = file.read()

doc = nlp(text)

word_counts = Counter([token.text.lower() for token in doc if not token.is_stop and not token.is_punct])

most_common_word, frequency = word_counts.most_common(1)[0]

print(f"Das häufigste Wort ist '{most_common_word}' mit einer Häufigkeit von {frequency}.")
print(word_counts.most_common(1))
```
